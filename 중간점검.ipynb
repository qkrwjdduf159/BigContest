```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_excel('/content/drive/MyDrive/빅콘테스트/2021 빅콘테스트_데이터분석분야_챔피언리그_수산Biz_문제데이터.xlsx')
val = pd.read_excel('/content/drive/MyDrive/빅콘테스트/2021 빅콘테스트_데이터분석분야_챔피언리그_수산Biz_자율평가데이터.xlsx')
test = pd.read_excel('/content/drive/MyDrive/빅콘테스트/2021 빅콘테스트_데이터분석분야_챔피언리그_수산Biz_평가데이터_update_210831.xlsx')
```

```python
train = train.rename(columns = {'REG_DATE':'기준일', 'P_TYPE':'제품구분','CTRY_1':'제조국','CTRY_2':'수출국',
                        'P_PURPOSE':'수입용도','CATEGORY_1':'중분류명 카테고리',
                        'CATEGORY_2':'어종','P_NAME':'상세어종','P_IMPORT_TYPE':'수입형태',
                        'P_PRICE':'target'})

val = val.rename(columns = {'REG_DATE':'기준일', 'P_TYPE':'제품구분','CTRY_1':'제조국','CTRY_2':'수출국',
                        'P_PURPOSE':'수입용도','CATEGORY_1':'중분류명 카테고리',
                        'CATEGORY_2':'어종','P_NAME':'상세어종','P_IMPORT_TYPE':'수입형태',
                        'P_PRICE':'target'})
```

```python
train_연어 = train[train['상세어종'] == '연어'].reset_index(drop = True)
val_연어 = val[val['상세어종'] == '연어'].reset_index(drop = True)

train_오징어 = train[train['상세어종'] == '오징어'].reset_index(drop = True)
val_오징어 = val[val['상세어종'] == '오징어'].reset_index(drop = True)

train_흰다리새우 = train[train['상세어종'] == '흰다리새우'].reset_index(drop = True)
val_흰다리새우 = val[val['상세어종'] == '흰다리새우'].reset_index(drop = True)
```

```python
df_연어 = pd.concat([train_연어, val_연어], axis = 0)
df_오징어 = pd.concat([train_오징어, val_오징어],axis = 0)
df_흰다리새우 = pd.concat([train_흰다리새우, val_흰다리새우], axis = 0)
```

```python
for i in ['어종','상세어종','어종.1','상세어종.1','어종.2','상세어종.2']:
  test[i] = test[i][0]
```

```python
test['week'] = test['일자'].dt.week
test['month'] = test['일자'].dt.month
```

```python
df_연어['month'] = df_연어['기준일'].dt.month
df_연어['week'] = df_연어['기준일'].dt.week

df_오징어['month'] = df_오징어['기준일'].dt.month
df_오징어['week'] = df_오징어['기준일'].dt.week

df_흰다리새우['month'] = df_흰다리새우['기준일'].dt.month
df_흰다리새우['week'] = df_흰다리새우['기준일'].dt.week
```

```python
# 모델을 위한 패키지 가져오기
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_predict, KFold
from vecstack import stacking

def rmse(y, y_pred):
  return np.sqrt(mean_squared_error(y, y_pred))

final_dict = {}
final_dict['연어'] = df_연어[['month','week','target']]
final_dict['오징어'] = df_오징어[['month','week','target']]
final_dict['흰다리새우'] = df_흰다리새우[['month','week','target']]


from sklearn.linear_model import LinearRegression
final_y_pred = {}

for idx, value in final_dict.items():
  X_train, y_train = final_dict[idx].drop('target', axis = 1), final_dict[idx]['target']
  X_test = test[['month','week']]

  # 모델링
  lgb = LGBMRegressor(random_state =42)
  xgb = XGBRegressor(random_state = 42, tree_method = 'gpu_hist')
  catboost = CatBoostRegressor(random_state =42, task_type = 'GPU', verbose = False)
  rfg = RandomForestRegressor(random_state = 42)
  ridge = Ridge(random_state = 42)
  lr = LinearRegression()
  gb = GradientBoostingRegressor(random_state = 42)
  lasso = Lasso(random_state = 42)
  
  models = [lgb,xgb, catboost, ridge, rfg,lr, gb, lasso]

  S_train, S_test = stacking(models, X_train, y_train, X_test, regression = True, mode = 'oof_pred_bag', metric  = rmse, n_folds =3, random_state = 42)

  lr = LinearRegression()
  lr.fit(S_train, y_train)
  y_pred = lr.predict(S_test)

  final_y_pred[idx] = y_pred
```
